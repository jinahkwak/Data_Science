```python
import pandas as pd
train = pd.read_csv('train.csv')

train.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## 성별별 생존자, 사망자 수 Count

```python
sns.catplot('Survived', data = train, hue = 'Sex', kind ='count')
```

## 계급별 생존자, 사망자 수 Count

```python
sns.catplot('Survived', data=train, hue='Pclass', kind='count')
```

## 승선한 항구별 생존자, 사망자 수 Count

```python
sns.catplot('Survived', data=train, hue='Embarked', kind='count')
```

## 나이별 생존자, 사망자 수 Count

```python
sns.catplot(data=train, x='Survived', hue='Age', kind='count')
```
